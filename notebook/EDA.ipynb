# MODEL TRAINING

#### 1.1 Import Data and Required Packages
##### Importing Pandas, Numpy, Matplotlib, Seaborn and Warings Library.

```python
# Basic Import
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
# Modeling - Classification
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import AdaBoostClassifier


# Boosting Libraries
from xgboost import XGBClassifier
from catboost import CatBoostClassifier
```

#### Show Top 5 Records

```python
df.head()
```

#### Encode diagnosis column

```python
df['diagnosis'] = df['diagnosis'].map({'B': 0, 'M': 1})
```

```python
df.head()
```

#### Preparing X and Y variables

```python
X = df.drop('diagnosis', axis=1)
y = df['diagnosis']
```

```python
X.head()
```

```python
# Scale features
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# separate dataset into train and test
# separate dataset into train and test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scaled,y,test_size=0.2,random_state=42)
X_train.shape, X_test.shape
```

#### Create an Evaluate Function to give all metrics after model Training

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

def evaluate_classification(true, predicted):
    acc = accuracy_score(true, predicted)
    prec = precision_score(true, predicted, average='binary') 
    rec = recall_score(true, predicted, average='binary')
    f1 = f1_score(true, predicted, average='binary')
    cm = confusion_matrix(true, predicted)
    return acc, prec, rec, f1, cm
```

```python
# Model dictionary
models = {
    "Logistic Regression": LogisticRegression(),
    "Random Forest": RandomForestClassifier(),
    "SVM": SVC(probability=True),
    "AdaBoost": AdaBoostClassifier(),
    "XGBoost": XGBClassifier(),
    "CatBoost": CatBoostClassifier(verbose=False)
}

# Evaluation loop
model_names = []
accuracy_scores = []

for name, model in models.items():
    try:
        model.fit(X_train, y_train)

        # Predictions
        y_train_pred = model.predict(X_train)
        y_test_pred = model.predict(X_test)

        # Train metrics
        acc_train, prec_train, rec_train, f1_train, cm_train = evaluate_classification(y_train, y_train_pred)

        # Test metrics
        acc_test, prec_test, rec_test, f1_test, cm_test = evaluate_classification(y_test, y_test_pred)

        print(f"🔍 {name}")
        model_names.append(name)
        accuracy_scores.append(acc_test)

        print("📊 Training Set Performance:")
        print(f"- Accuracy: {acc_train:.4f}")
        print(f"- Precision: {prec_train:.4f}")
        print(f"- Recall: {rec_train:.4f}")
        print(f"- F1 Score: {f1_train:.4f}")
        print(f"- Confusion Matrix:\n{cm_train}")
        print("----------------------------------------")

        print("📊 Test Set Performance:")
        print(f"- Accuracy: {acc_test:.4f}")
        print(f"- Precision: {prec_test:.4f}")
        print(f"- Recall: {rec_test:.4f}")
        print(f"- F1 Score: {f1_test:.4f}")
        print(f"- Confusion Matrix:\n{cm_test}")
        print("=" * 50 + "\n")

    except Exception as e:
        print(f"❌ {name} failed: {e}")
        print("=" * 50 + "\n")
```

```python
# Model dictionary
models = {
    "Logistic Regression": LogisticRegression(),
    "Random Forest": RandomForestClassifier(),
    "SVM": SVC(probability=True),
    "AdaBoost": AdaBoostClassifier(),
    "XGBoost": XGBClassifier(),
    "CatBoost": CatBoostClassifier(verbose=False)
}

# Evaluation loop
model_names = []
accuracy_scores = []

with open("model_evaluation.txt", "w", encoding="utf-8") as f:
    for name, model in models.items():
        try:
            model.fit(X_train, y_train)

            # Predictions
            y_train_pred = model.predict(X_train)
            y_test_pred = model.predict(X_test)

            # Train metrics
            acc_train, prec_train, rec_train, f1_train, cm_train = evaluate_classification(y_train, y_train_pred)

            # Test metrics
            acc_test, prec_test, rec_test, f1_test, cm_test = evaluate_classification(y_test, y_test_pred)

            model_names.append(name)
            accuracy_scores.append(acc_test)

            f.write(f"🔍 {name}\n")
            f.write("Training Set Performance:\n")
            f.write(f"- Accuracy: {acc_train:.4f}\n")
            f.write(f"- Precision: {prec_train:.4f}\n")
            f.write(f"- Recall: {rec_train:.4f}\n")
            f.write(f"- F1 Score: {f1_train:.4f}\n")
            f.write(f"- Confusion Matrix:\n{cm_train}\n")
            f.write("----------------------------------------\n")
            f.write("Test Set Performance:\n")
            f.write(f"- Accuracy: {acc_test:.4f}\n")
            f.write(f"- Precision: {prec_test:.4f}\n")
            f.write(f"- Recall: {rec_test:.4f}\n")
            f.write(f"- F1 Score: {f1_test:.4f}\n")
            f.write(f"- Confusion Matrix:\n{cm_test}\n")
            f.write("=" * 80 + "\n\n")

        except Exception as e:
            f.write(f"❌ {name} failed: {e}\n")
            f.write("=" * 80 + "\n\n")
```
